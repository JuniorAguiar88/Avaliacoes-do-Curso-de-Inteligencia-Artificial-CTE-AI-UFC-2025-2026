## Dataset de Correção de Bugs

Este dataset foi utilizado na pesquisa publicada em [1] e consiste em um conjunto de informações sobre tarefas de correção de bugs no projeto [Apache Hadoop](https://hadoop.apache.org/), coletadas ao longo de um período de `10` anos. Além disso, o dataset foi manualmente rotulado para indicar se o bug registrado é ou não um bug de tratamento de exceção.

```bash
[1] da Silva, A.J.A., Vieira, R.G., Mesquita, D.P.P. et al. Towards automatic labeling of exception handling bugs: A case study of 10 years bug-fixing in Apache Hadoop. Empir Software Eng 29, 85 (2024). https://doi.org/10.1007/s10664-024-10494-0
```

A arquitetura do Hadoop compreende quatro componentes principais:

- `HADOOP` (ou core): que fornece o pacote de utilitários para suportar outros módulos do Hadoop;
- `MAPREDUCE`: um modelo de programação para armazenamento e processamento de dados;
- `HDFS` (*Hadoop Distributed File System*): um sistema de arquivos distribuído que roda em clusters projetados para armazenar arquivos muito grandes e fornecer acesso de alta taxa de transferência; e
- `YARN` (*Yet Another Resource Negotiator*): um framework para agendamento/monitoramento de tarefas e gerenciamento de recursos do cluster.

### Questões 1, 2 e 3 - K-means

O artigo menciona que bugs de Exceção (EH) tendem a ser mais discutidos (mais comentários) e levam mais tempo para serem corrigidos. Vamos usar Clustering para ver se o algoritmo descobre esses grupos "difíceis" sem saber o rótulo.

1. Selecione as features de "Interação e Tempo": NoComments, NoCommits, ResolutionTime.

1. Aplique StandardScaler para normalizar.

1. Rode o K-Means com n_clusters=3, random_state=42, n_init=10.

1. Analise os centróides. O Cluster que possui, em média, o maior tempo de correção e o maior número de comentários é o nosso "Cluster Crítico"

````python
#1. Selecione as features de "Interação e Tempo": NoComments, NoCommits,
#ResolutionTime:

# 1. Primeiramente precisamos upar a tabela:
import pandas as pd

#from google.colab import files
#uploaded = files.upload()

# 2. Após upar a tabela precisamos corrigir o separador da tabela, pois antes
#dos cálculos, pois verifiquei a estrutura do dataset e identifiquei que o
#arquivo original usava ";" como separador — corrigí isso no carregamento, para
#garantir que as 54 colunas pudessem ser lidas adequadamente.
# Através da lib "pandas" o método pd.read_csv() é responsável por ler arquivos
#no formato .csv que foi upado anteriormente e possui o nome de arquivo:
#labeled-hadoop-bug-fix-dataset.csv que a pós o upload anterior ficou disponível
#na célula para se trabalhado;
#sep=';' é o parâmetro que define e diz pro interpretador que o caractere
#separador de colunas é ;
df = pd.read_csv('labeled-hadoop-bug-fix-dataset.csv', sep=';')

print("Colunas disponíveise seus tipos:")
# 3. vamos verificar as colunas e os seus tipos no dataframe

for col in df.columns:
    print(f"{col:<25} → {df[col].dtype}")

print("A coluna ResolutionTime não existe!!")

print("Vamos criar a coluna ResolutionTime no dataset a partir diferença das colunas CreationDate e ResolutionDate")

# 4. Criando a coluna ResolutionTime a partir das colunas
df['ResolutionTime'] = (
    pd.to_datetime(df['ResolutionDate'], errors='coerce') -
    pd.to_datetime(df['CreationDate'], errors='coerce')
).dt.total_seconds() / (24 * 60 * 60) #24 HORAS, 60 MINUTOS E 60 SEGUNDOS PREPARANDO
#A COLUNA PARA SE TER FVALORES DE TEMPO/DATA
#OBS. IMPORTANTE: A coluna ResolutionTime foi criada como ResolutionDate
#(tempo de resolução) – CreationDate (data de criação) porque essa é a única
#definição logicamente válida para “tempo de resolução - ResolutionTime”

# 5. Imprimir as três features pedidas e seus tipos
for col in ['NoComments', 'NoCommits', 'ResolutionTime']:
    print(f"{col:<15} → {df[col].dtype}")

#6. Selecionando as features de "Interação e tempo"
X = df[['NoComments', 'NoCommits', 'ResolutionTime']]
print(f"As features de interação e tempo {X.columns.tolist()} foram selecionadas.")

````

````python
#2. Aplique StandardScaler para normalizar:

# Se não normalizarmos, features com valores maiores (como ResolutionTime)
# dominarão a distância no clustering — o que distorce os resultados.


#Após a soleção das features 'NoComments', 'NoCommits', 'ResolutionTime' e para
#que os resultados com valores maiores de ResolutionTime e que não
#domine a distância [(0.01 -> 2785.90) muito distantes] do clustering distorcendo os resultados

from sklearn.preprocessing import StandardScaler
#StandardScaler normaliza as features

# 1. Verificar as escalas reais das features antes da normalização
print("Escala das features antes da normalização:")

for col in ['NoComments', 'NoCommits', 'ResolutionTime']:
    min_val = df[col].min()
    max_val = df[col].max()
    mean_val = df[col].mean()
    std_val = df[col].std()
    print(f"{col:<15} → min: {min_val:>8.2f} | max: {max_val:>10.2f} | média: {mean_val:>8.2f} | std: {std_val:>8.2f}\n")
# As três features têm escalas muito diferentes:
# - NoComments: geralmente pequeno ( 0 a 189.00)
# - NoCommits: também pequeno (0 a 6.0)
# - ResolutionTime: pode ser grande (0.1 a 2785.90 dias)

# 2. Instancia o normalizador
scaler = StandardScaler()

# 3. Ajustar o scaler aos dados (calcula média e std) e aplica a transformação
X_scaled = scaler.fit_transform(X)

# 4. Converter de volta para DataFrame para manter legibilidade (nomes das colunas, índice)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)

print("Médias após normalização (devem ser ≈ 0):")
print(X_scaled.mean().round(6))
print("\nDesvios padrão após normalização (devem ser = 1):")
print(X_scaled.std().round(6))
#transforma cada coluna para média = 0 e desvio padrão = 1.
````

````python
#3. Rode o K-Means com n_clusters=3, random_state=42, n_init=10.

from sklearn.cluster import KMeans

# 1. Aplicar K-Means com os parâmetros solicitados
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# 2. Treinar o modelo nos dados normalizados
kmeans.fit(X_scaled)

# 3. Atribuir os rótulos de cluster ao dataset original
df['Cluster'] = kmeans.labels_

print(f"• Número de clusters: {kmeans.n_clusters}")
print(f"• Rótulos atribuídos: {sorted(df['Cluster'].unique())}")
print(f"• Inércia (soma das distâncias quadradas aos centróides): {kmeans.inertia_:.2f}")
````

````python
#4. Analise os centróides. O Cluster que possui, em média, o maior tempo de correção e o maior número de comentários é o nosso "Cluster Crítico"

# 1. Precisamos entender o comportamento de cada cluster e agrupar os dados pelo rótulo do K-Means e calcular a média das três features-chave:
# - NoComments: indicando quanto o bug foi discutido
# - NoCommits: "sugere" o esforço de implementação
# - ResolutionTime: revela quanto tempo levou para resolver
# Arredondando para 2 casas para facilitar a leitura
means = df.groupby('Cluster')[['NoComments', 'NoCommits', 'ResolutionTime']].mean().round(2)

# 2. Mostrar as médias para comparar qual cluster tem mais interação e qual demora mais
print("Médias por cluster:")
print(means)

# 3. identificar o "Cluster Crítico".
# Pelo enunciado, creio que ele deve ser o que tem MAIOR tempo de correção (ResolutionTime)
# E também o MAIOR número de comentários (NoComments).
# Então verificamos se o cluster com maior ResolutionTime é o mesmo com maior NoComments?!
max_rt = means['ResolutionTime'].idxmax()   # Qual cluster tem mais tempo?
max_nc = means['NoComments'].idxmax()       # Qual cluster tem mais comentários?

# 4. verificar se o mesmo cluster lidera nas duas métricas.
if max_rt == max_nc:
    critical = max_rt
    print(f"\nCluster Crítico: {critical} (maior em ambas as métricas)")
# Mas... e se não for o mesmo? Isso pode acontecer — por exemplo, um cluster com muitos comentários mas rápido,
# e outro lento, mas com pouca discussão. Nesse caso, preciso de um critério justo.
# Optei por padronizar as médias (z-score dentro dos clusters) e somar:
# assim, tempo e comentários têm o mesmo peso — evito que a escala residual distorça a decisão.
else:
    print(f"\nEmpate: Cluster {max_rt} tem mais tempo, Cluster {max_nc} tem mais comentários.")
    # Normalizo internamente entre os clusters (não uso o scaler global, pois só há 3 pontos)
    rt_norm = (means['ResolutionTime'] - means['ResolutionTime'].min()) / means['ResolutionTime'].std()
    nc_norm = (means['NoComments'] - means['NoComments'].min()) / means['NoComments'].std()
    means['Score'] = rt_norm + nc_norm  # Quanto maior, mais "crítico"
    critical = means['Score'].idxmax()
    print(f"Cluster Crítico (maior soma padronizada): {critical}")

# 5. reportar os valores reais do cluster crítico — em dias e comentários.
print(f"   → Tempo médio: {means.loc[critical, 'ResolutionTime']:.2f} dias")
print(f"   → Comentários médios: {means.loc[critical, 'NoComments']:.2f}")
````

### Questões 4, 5

Entender a redundância nos dados de métricas do Git. Contexto: Temos muitas métricas correlacionadas (NoAuthors, NoCommitters, NoCommits, etc.). Queremos reduzir isso para visualizar o espaço.

Instruções de Código:

1. Selecione as colunas: 'NoComments', 'NoCommits', 'NoAttachments', 'NoAttachedPatches', 'NoAuthors', 'SrcAddLines', 'SrcDelLines', 'ResolutionTime'.

1. Normalize os dados com StandardScaler.

1. Aplique PCA (Principal Component Analysis) com random_state=42.

1. Verifique a Explained Variance Ratio (razão da variância explicada).

````python

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Selecionar as features de métricas do Git e interação
git_features = [
    'NoComments',
    'NoCommits',
    'NoAttachments',
    'NoAttachedPatches',
    'NoAuthors',
    'SrcAddLines',
    'SrcDelLines',
    'ResolutionTime'
]

# Selecionar efetivamente
X_git = df[git_features]

print("Colunas selecionadas:")
print(X_git.columns.tolist())

# Instanciar e aplicar StandardScaler
scaler_git = StandardScaler()
X_git_scaled = scaler_git.fit_transform(X_git)

## Converter para DataFrame (mantendo nomes das colunas)
X_git_scaled = pd.DataFrame(X_git_scaled, columns=X_git.columns, index=X_git.index)

print("Normalização concluída.")
print("Médias (devem ser ~0):")
print(X_git_scaled.mean().round(6))
print("\nDesvios padrão (devem ser = 1):")
print(X_git_scaled.std().round(6))

# Aplicar PCA com random_state para reprodutibilidade
pca = PCA(random_state=42)

# Ajustar o PCA e transformar os dados
X_pca = pca.fit_transform(X_git_scaled)

# Converter para DataFrame com nomes dos componentes
n_components = X_pca.shape[1]  # = 8 (mesmo número de features originais)
X_pca_df = pd.DataFrame(
    X_pca,
    columns=[f'PC{i+1}' for i in range(n_components)],
    index=X_git.index
)


print(f"\n• Número de componentes: {n_components}")
print(f"• Variância explicada por componente: {pca.explained_variance_ratio_.round(4)}")
print(f"• Variância total explicada (todos os PCs): {pca.explained_variance_ratio_.sum():.4f}")

# Exibir a variância explicada por componente
evr = pca.explained_variance_ratio_
cumsum = evr.cumsum()

print("Explained Variance Ratio (variância explicada por componente):")
for i, (ratio, cum) in enumerate(zip(evr, cumsum), 1):
    print(f"PC{i:>2}: {ratio:.4f}  →  acumulado: {cum:.4f}")

print(f"\n Total explicado com:")
print(f"  • 1 componente: {cumsum[0]:.1%}")
print(f"  • 2 componentes: {cumsum[1]:.1%}")
print(f"  • 3 componentes: {cumsum[2]:.1%}")
print(f"  • 4 componentes: {cumsum[3]:.1%}")
````

````python
#6. Observando a região da "Cauda Longa" do gráfico KDE (tempo de correção > 50 dias), qual comportamento é observado?

import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar bugs resolvidos
resolved = df[df['ResolutionTime'] > 0]['ResolutionTime']

# Plotar KDE com bw_adjust=0.5
plt.figure(figsize=(8, 4.5))
sns.kdeplot(resolved, fill=True, alpha=0.6, color='steelblue', bw_adjust=0.5)

# Linha de 50 dias
plt.axvline(50, color='red', linestyle='--', linewidth=1.5, label='50 dias')

# Percentual na cauda
pct_long = (resolved > 50).mean() * 100
plt.text(55, plt.ylim()[1] * 0.85,
         f'{pct_long:.1f}% dos bugs\nlevam > 50 dias',
         fontsize=10, color='red',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="white", edgecolor="red", alpha=0.8))

plt.xlim(0, 300)
plt.xlabel('Tempo de correção (dias)')
plt.ylabel('Densidade')
plt.title('KDE de ResolutionTime')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````
